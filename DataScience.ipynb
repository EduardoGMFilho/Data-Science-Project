```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregando os dados
url = 'https://raw.githubusercontent.com/TailUFPB/processo2fase/refs/heads/master/2025/data/california.csv'
df = pd.read_csv(url)

# VisualizaÃ§Ã£o inicial
print(df.head())
print("\nInformaÃ§Ãµes gerais:")
df.info()

# EstatÃ­sticas descritivas
print("\nEstatÃ­sticas bÃ¡sicas:")
print(df.describe())

# Valores Ãºnicos por coluna
print("\nValores Ãºnicos por coluna:")
print(df.nunique())

# Verificando valores nulos
print("\nValores nulos por coluna:")
print(df.isnull().sum())

# Tratamento de dados ausentes (total_bedrooms)
df['total_bedrooms'].fillna(df['total_bedrooms'].median(), inplace=True)

# Histogramas das variÃ¡veis numÃ©ricas
num_cols = df.select_dtypes(include='number').columns

#LaÃ§o for que percorre as colunas para plot
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'DistribuiÃ§Ã£o da variÃ¡vel: {col}')
    plt.xlabel(col)
    plt.ylabel('FrequÃªncia')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Boxplots para detecÃ§Ã£o de outliers
for col in num_cols:
    plt.figure(figsize=(8, 2))
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot da variÃ¡vel: {col}')
    plt.tight_layout()
    plt.show()

# GrÃ¡fico de barras para variÃ¡vel categÃ³rica 'ocean_proximity'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='ocean_proximity')
plt.title('FrequÃªncia de cada categoria em ocean_proximity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# GrÃ¡fico de dispersÃ£o entre renda e valor de casa
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='median_income', y='median_house_value', alpha=0.3)
plt.title('RelaÃ§Ã£o entre renda mediana e valor da casa')
plt.xlabel('Renda Mediana')
plt.ylabel('Valor Mediano da Casa')
plt.tight_layout()
plt.show()
```

# AnÃ¡lise dos dados


* As variÃ¡veis como total_rooms, population e median_house_value apresentaram grande dispersÃ£o, com valores extremos.
* Tratamento de Dados NecessÃ¡rio: Foi identificado que a coluna total_bedrooms possui valores nulos. O tratamento adequado foi o preenchimento com a **mediana**, pois os dados apresentam possÃ­vel assimetria.


*  Destaco tambÃ©m uma relaÃ§Ã£o positiva entre median_income e median_house_valueItem da lista, isso implica que Ã  medida que a renda mediana aumenta em uma regiÃ£o, o valor mediano das casas tambÃ©m tende a aumentar.





# Limpeza dos Dados

Segue abaixo algumas alteraÃ§Ãµes realizadas para melhorar a visualizaÃ§Ã£o dos dados:


1. Trata valores nulos com a mediana

2.  Corrige o tipo da coluna median_income

1. Corrige o tipo da coluna median_income
2. Padroniza os valores categÃ³ricos e os nomes das colunas

1. Apresenta grÃ¡ficos para visualizaÃ§Ã£o de outliers














```python
# Verificar tipos de dados
print("Tipos de dados antes da limpeza:")
print(df.dtypes)

# Corrigir formato da coluna median_income (trocar vÃ­rgula por ponto e converter para float)
df['median_income'] = df['median_income'].astype(str).str.replace(',', '.') #MÃ©todo replace para conversÃ£o
df['median_income'] = df['median_income'].astype(float)

# Verificar e tratar valores ausentes
print("\nValores ausentes por coluna:")
print(df.isnull().sum())

# Preencher valores ausentes em total_bedrooms com a mediana
df['total_bedrooms'].fillna(df['total_bedrooms'].median(), inplace=True)

# Remover linhas duplicadas
print("\nNÃºmero de linhas duplicadas:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

# Verificar colunas categÃ³ricas e padronizar
print("\nValores Ãºnicos em 'ocean_proximity':")
print(df['ocean_proximity'].unique())

# Padronizar os valores da coluna ocean_proximity
df['ocean_proximity'] = df['ocean_proximity'].str.strip().str.upper()

# Padronizar nomes das colunas
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')

# Reordenar colunas (opcional)
cols = ['longitude', 'latitude'] + [col for col in df.columns if col not in ['longitude', 'latitude']]
df = df[cols]

# Verificar possÃ­veis outliers com boxplots (exemplo para 3 variÃ¡veis)
for col in ['median_income', 'median_house_value', 'housing_median_age']:
    plt.figure(figsize=(6, 2))
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot de {col}')
    plt.tight_layout()
    plt.show()

# Resultado final
print("\nDataset apÃ³s limpeza:")
print(df.info())
print(df.head())
```

# TÃ©cnica de Feature Engineering

Ã‰ comum nos depararmos com um dataset cujos atributos nÃ£o esclareÃ§am com efetividade o comportamento de um dado fenÃ´meno. Desta forma, manipulei atributos existentes e assim defini novos atributos, de acordo com os critÃ©rio que estabeleci.


O que foi feito ?



* CriaÃ§Ã£o de variÃ¡veis mais informativas e com poder explicativo
*  Reduz a complexidade do dataset sem perda de informaÃ§Ã£o.

* Torna os dados mais interpretÃ¡veis e modelÃ¡veis.
* Finaliza com um grÃ¡fico da faixa de renda, uma variÃ¡vel importante.







```python
# ForÃ§a a conversÃ£o de todas as colunas
cols_to_convert = ['total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']

# LaÃ§o para percorrer as colunas
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Se tiver string invÃ¡lida, vira NaN

# Pessoas por casa
df['people_per_household'] = df['population'] / df['households']

# Quartos por pessoa
df['rooms_per_person'] = df['total_rooms'] / df['population']

# Quartos por casa
df['rooms_per_household'] = df['total_rooms'] / df['households']

# RelaÃ§Ã£o entre quartos e dormitÃ³rios
df['bedroom_ratio'] = df['total_bedrooms'] / df['total_rooms']

# Indicador binÃ¡rio para localizaÃ§Ã£o costeira
df['is_coastal'] = df['ocean_proximity'].apply(lambda x: 1 if 'NEAR OCEAN' in x or 'ISLAND' in x else 0)

# Faixa de renda categÃ³rica
df['income_level'] = pd.cut(
    df['median_income'],
    bins=[0, 2, 4, 6, 8, df['median_income'].max()],
    labels=['muito_baixa', 'baixa', 'mÃ©dia', 'alta', 'muito_alta']
)

# Verificando os novos atributos
print("\nNovos atributos criados:")
print(df[['people_per_household', 'rooms_per_person', 'rooms_per_household',
          'bedroom_ratio', 'is_coastal', 'income_level']].head())

# VisualizaÃ§Ã£o rÃ¡pida
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='income_level', palette='Blues')
plt.title('DistribuiÃ§Ã£o das Faixas de Renda')
plt.tight_layout()
plt.show()
```

# Pensando em Futuros Projetos

Os algoritmos de Machine Learning sÃ³ funcionam com **dados numÃ©ricos**, com isso os atributos categÃ³ricos foram transformados em valores numÃ©ricos por meio de **One-Hot Encoding**.

 Esse processo garante que variÃ¡veis como ocean_proximity e income_level sejam representadas de forma adequada no modelo.

**One-Hot Encoding (binÃ¡rio por categoria)**


* IdentificaÃ§Ã£o dados categÃ³ricos
* Cria uma nova coluna para cada categoria com 0 ou 1.


```python
# Garantir que a coluna income_level esteja como string
df['income_level'] = df['income_level'].astype(str)

# AplicaÃ§Ã£o One-Hot Encoding nas colunas categÃ³ricas
df = pd.get_dummies(df, columns=['ocean_proximity', 'income_level'], prefix=['ocean', 'income'])

# VerificaÃ§Ã£o das novas colunas adicionadas
print("Colunas do dataset apÃ³s codificaÃ§Ã£o:")
print(df.columns)

# 4. VerificaÃ§Ã£o da amostra dos dados
print("\nExemplo de linhas com novas colunas codificadas:")
print(df.head())
```
