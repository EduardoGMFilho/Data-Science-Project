```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregando os dados
url = 'https://raw.githubusercontent.com/TailUFPB/processo2fase/refs/heads/master/2025/data/california.csv'
df = pd.read_csv(url)

# Visualização inicial
print(df.head())
print("\nInformações gerais:")
df.info()

# Estatísticas descritivas
print("\nEstatísticas básicas:")
print(df.describe())

# Valores únicos por coluna
print("\nValores únicos por coluna:")
print(df.nunique())

# Verificando valores nulos
print("\nValores nulos por coluna:")
print(df.isnull().sum())

# Tratamento de dados ausentes (total_bedrooms)
df['total_bedrooms'].fillna(df['total_bedrooms'].median(), inplace=True)

# Histogramas das variáveis numéricas
num_cols = df.select_dtypes(include='number').columns

#Laço for que percorre as colunas para plot
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribuição da variável: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Boxplots para detecção de outliers
for col in num_cols:
    plt.figure(figsize=(8, 2))
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot da variável: {col}')
    plt.tight_layout()
    plt.show()

# Gráfico de barras para variável categórica 'ocean_proximity'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='ocean_proximity')
plt.title('Frequência de cada categoria em ocean_proximity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico de dispersão entre renda e valor de casa
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='median_income', y='median_house_value', alpha=0.3)
plt.title('Relação entre renda mediana e valor da casa')
plt.xlabel('Renda Mediana')
plt.ylabel('Valor Mediano da Casa')
plt.tight_layout()
plt.show()
```

# Análise dos dados


* As variáveis como total_rooms, population e median_house_value apresentaram grande dispersão, com valores extremos.
* Tratamento de Dados Necessário: Foi identificado que a coluna total_bedrooms possui valores nulos. O tratamento adequado foi o preenchimento com a **mediana**, pois os dados apresentam possível assimetria.


*  Destaco também uma relação positiva entre median_income e median_house_valueItem da lista, isso implica que à medida que a renda mediana aumenta em uma região, o valor mediano das casas também tende a aumentar.





# Limpeza dos Dados

Segue abaixo algumas alterações realizadas para melhorar a visualização dos dados:


1. Trata valores nulos com a mediana

2.  Corrige o tipo da coluna median_income

1. Corrige o tipo da coluna median_income
2. Padroniza os valores categóricos e os nomes das colunas

1. Apresenta gráficos para visualização de outliers














```python
# Verificar tipos de dados
print("Tipos de dados antes da limpeza:")
print(df.dtypes)

# Corrigir formato da coluna median_income (trocar vírgula por ponto e converter para float)
df['median_income'] = df['median_income'].astype(str).str.replace(',', '.') #Método replace para conversão
df['median_income'] = df['median_income'].astype(float)

# Verificar e tratar valores ausentes
print("\nValores ausentes por coluna:")
print(df.isnull().sum())

# Preencher valores ausentes em total_bedrooms com a mediana
df['total_bedrooms'].fillna(df['total_bedrooms'].median(), inplace=True)

# Remover linhas duplicadas
print("\nNúmero de linhas duplicadas:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

# Verificar colunas categóricas e padronizar
print("\nValores únicos em 'ocean_proximity':")
print(df['ocean_proximity'].unique())

# Padronizar os valores da coluna ocean_proximity
df['ocean_proximity'] = df['ocean_proximity'].str.strip().str.upper()

# Padronizar nomes das colunas
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')

# Reordenar colunas (opcional)
cols = ['longitude', 'latitude'] + [col for col in df.columns if col not in ['longitude', 'latitude']]
df = df[cols]

# Verificar possíveis outliers com boxplots (exemplo para 3 variáveis)
for col in ['median_income', 'median_house_value', 'housing_median_age']:
    plt.figure(figsize=(6, 2))
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot de {col}')
    plt.tight_layout()
    plt.show()

# Resultado final
print("\nDataset após limpeza:")
print(df.info())
print(df.head())
```

# Técnica de Feature Engineering

É comum nos depararmos com um dataset cujos atributos não esclareçam com efetividade o comportamento de um dado fenômeno. Desta forma, manipulei atributos existentes e assim defini novos atributos, de acordo com os critério que estabeleci.


O que foi feito ?



* Criação de variáveis mais informativas e com poder explicativo
*  Reduz a complexidade do dataset sem perda de informação.

* Torna os dados mais interpretáveis e modeláveis.
* Finaliza com um gráfico da faixa de renda, uma variável importante.







```python
# Força a conversão de todas as colunas
cols_to_convert = ['total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']

# Laço para percorrer as colunas
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Se tiver string inválida, vira NaN

# Pessoas por casa
df['people_per_household'] = df['population'] / df['households']

# Quartos por pessoa
df['rooms_per_person'] = df['total_rooms'] / df['population']

# Quartos por casa
df['rooms_per_household'] = df['total_rooms'] / df['households']

# Relação entre quartos e dormitórios
df['bedroom_ratio'] = df['total_bedrooms'] / df['total_rooms']

# Indicador binário para localização costeira
df['is_coastal'] = df['ocean_proximity'].apply(lambda x: 1 if 'NEAR OCEAN' in x or 'ISLAND' in x else 0)

# Faixa de renda categórica
df['income_level'] = pd.cut(
    df['median_income'],
    bins=[0, 2, 4, 6, 8, df['median_income'].max()],
    labels=['muito_baixa', 'baixa', 'média', 'alta', 'muito_alta']
)

# Verificando os novos atributos
print("\nNovos atributos criados:")
print(df[['people_per_household', 'rooms_per_person', 'rooms_per_household',
          'bedroom_ratio', 'is_coastal', 'income_level']].head())

# Visualização rápida
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='income_level', palette='Blues')
plt.title('Distribuição das Faixas de Renda')
plt.tight_layout()
plt.show()
```

# Pensando em Futuros Projetos

Os algoritmos de Machine Learning só funcionam com **dados numéricos**, com isso os atributos categóricos foram transformados em valores numéricos por meio de **One-Hot Encoding**.

 Esse processo garante que variáveis como ocean_proximity e income_level sejam representadas de forma adequada no modelo.

**One-Hot Encoding (binário por categoria)**


* Identificação dados categóricos
* Cria uma nova coluna para cada categoria com 0 ou 1.


```python
# Garantir que a coluna income_level esteja como string
df['income_level'] = df['income_level'].astype(str)

# Aplicação One-Hot Encoding nas colunas categóricas
df = pd.get_dummies(df, columns=['ocean_proximity', 'income_level'], prefix=['ocean', 'income'])

# Verificação das novas colunas adicionadas
print("Colunas do dataset após codificação:")
print(df.columns)

# 4. Verificação da amostra dos dados
print("\nExemplo de linhas com novas colunas codificadas:")
print(df.head())
```
